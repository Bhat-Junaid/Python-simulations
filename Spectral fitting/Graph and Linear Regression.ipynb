```python
import numpy as np
import matplotlib.pyplot as plt

# Loading the data files 
exp_spectra = np.loadtxt(r"C:\Academics\Numerical Methods\spectre_1.txt")     # Original Function values
theo_spectra_w1 = np.loadtxt(r"C:\Academics\Numerical Methods\theo_wi1_spectre_1.txt")  #  unweighted values
theo_spectra_w2 = np.loadtxt(r"C:\Academics\Numerical Methods\theo_wi2_spectre_1.txt")  # weighted values
omega = np.loadtxt(r"C:\Academics\Numerical Methods\omega_values.txt")            # Wavenumber values

# Plotting the data
plt.figure()
plt.scatter(omega, exp_spectra, facecolor = "none", edgecolor = "b", label="expt")
plt.plot(omega, theo_spectra_w1, "r-.", label = "calc. W=1")
plt.plot(omega, theo_spectra_w2, color = "orange" ,linestyle = "--", label = "calc. W=$F_{exp}^2$")
plt.xlabel("Wavenumber, $cm^{-1}$")
plt.ylabel("Intensity")
plt.title("Spectrum at Pressure = 1 atm")
plt.legend()
plt.show()
```
